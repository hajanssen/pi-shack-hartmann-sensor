```python
# time compleate exicution 

# Import of Libarasy
import matplotlib.pyplot as plt
import numpy as np
from simulation import Shak, getRange, gauss, gauss2D

# Generate Data 

# image details 
x_pos = np.linspace(10,90,6)
y_pos = np.linspace(10,90,6)

X,Y = np.meshgrid(x_pos,y_pos)

res = (480,640)
res = (480,640)
im_range_x=(0,100)


# Generate Image 
image = Shak(X, Y, res, 
            im_range_x, 
            im_range_x)

image2 = Shak(X-3, Y+3, res, 
            im_range_x, 
            im_range_x)

uint8 = lambda x : ((x/np.max(x) * 255)).astype(np.uint8)

# cast to uint8
img =  uint8(image)
img2 =  uint8(image2)


# Plot
# c = plt.imshow(img)
# plt.colorbar(c)
# plt.title("Simulated Reference")
# plt.show()

# c = plt.imshow(img2)
# plt.colorbar(c)
# plt.title("Simulated Measurment")
# plt.show()

# extract preak points 

def GetSeperation(img):
    from skimage.measure import label
    from skimage.feature import peak_local_max as peak
    from skimage.segmentation import expand_labels
    # get peak position of spots
    spots = peak(img, min_distance=10)

    # reshap peak point in x- and y-array
    x_pos = np.array([x[0] for x in spots])
    y_pos = np.array([x[1] for x in spots])

    #make image with "1" as peak, evrsthin els "0"
    img_mask = np.zeros_like(img)
    img_mask[x_pos,y_pos] = 1

    seg1 = label(img_mask)
    
    # [note] function uses very slow eucleadian distance implementatoin
    # cv has very fast one 
    expanded = expand_labels(seg1, distance=100)
    return expanded

def GetSeperationCV(img):
    import numpy as np
    from cv2 import distanceTransformWithLabels, DIST_L2
    from skimage.feature import peak_local_max as peak
    
    # get peak position of spots
    spots = peak(img, min_distance=10)

    # reshap peak point in x- and y-array
    x_pos = np.array([x[0] for x in spots])
    y_pos = np.array([x[1] for x in spots])

    #make image with "1" as peak, evrsthin els "0"
    img_mask = np.zeros_like(img, dtype=np.uint8)
    img_mask[x_pos,y_pos] = 1
    
    n, expanded =  distanceTransformWithLabels(np.uint8(img_mask == 0), DIST_L2,3)
    return expanded

def getMomentum(img_lables_in,img_ori):
    
    img_lables = np.copy(img_lables_in) + 1
    img_ori = np.copy(img_ori)
    
    x = np.zeros(np.max(img_lables)-1)
    y = np.zeros(np.max(img_lables)-1)
    #disk = morphology.disk(10)
    for i in range(2,np.max(img_lables)+1):
        mask = img_lables == i
        M = measure.moments(img_ori * mask)
        x[i-2] = M[0, 1] / M[0, 0]
        y[i-2] = M[1, 0] / M[0, 0]
        
    return x, y

def getMomentumCV(img_lables_in,img_ori):
    from cv2 import moments
    
    img_lables = np.copy(img_lables_in) + 1

    x = np.zeros(np.max(img_lables)-1)
    y = np.zeros(np.max(img_lables)-1)
    
    # [note] this loop could be paralised
    for i in range(2,np.max(img_lables)+1):
        mask = img_lables != i
        img_slice = np.copy(img_ori).astype(np.uint8)
        img_slice[mask] = 0 
    
        M = moments(img_slice)
        x[i-2] = M["m01"] / M["m00"]
        y[i-2] = M["m10"] / M["m00"]

    return y, x

def getSubFrameIndex(img):
    x_list = np.any(img,axis=1)
    x1 = np.argmax(x_list)
    x2 = np.argmax(np.invert(x_list[x1:-1])) + x1

    y_list = np.any(img,axis=0)
    y1 = np.argmax(y_list)
    y2 = np.argmax(np.invert(y_list[y1:-1])) + y1
    
    # if one "true" section has no "Fals" section it its right
    if x1 == x2:
        x2 = img.shape[0]    
    if y1 == y2:
        y2 = img.shape[1]

    return x1,x2,y1,y2


def getMomentumDIY(img_lables_in,img_ori, ploting = False):
    from cv2 import moments
    import matplotlib.patches as patches 
    
    img_lables = np.copy(img_lables_in) + 1

    x = np.zeros(np.max(img_lables)-1)
    y = np.zeros(np.max(img_lables)-1)
    
    # [note] this loop could be paralised
    
    x_len = np.shape(img)[0]
    y_len = np.shape(img)[1]
    
    for i in range(2,np.max(img_lables)+1):
        
         
        mask = img_lables == i
        img_slice = np.copy(img_ori).astype(np.uint8)
        
        
        if ploting:
            fig,axs = plt.subplots(1,3, dpi=200)
            axs[0].imshow(img_slice)
            axs[0].imshow(img_slice)
            axs[0].set_title(str(i))  
            axs[1].imshow(mask)
  
        
        # get lages recrangular around True values only asses pixels of interest
        x1,x2,y1,y2 = getSubFrameIndex(mask)
        
        mask = mask[x1:x2,y1:y2]
        img_slice = img_slice[x1:x2,y1:y2]
        
        
        img_slice[np.invert(mask)] = 0
        if ploting:
            print(x1,x2,y1,y2)
            axs[2].imshow(img_slice)
            axs[0].plot([y1,y2,y2,y1,y1],[x1,x1,x2,x2,x1])
            plt.show()
        
        M = moments(img_slice)

        x[i-2] = M["m01"] / M["m00"] + x1
        y[i-2] = M["m10"] / M["m00"] + y1
                
    return y, x


def getPartner(img1_x,img1_y,img2_x,img2_y):

    # find coresponding points in images, by looking for clooses on s
    partner = [];
    for i in range(0,len(img1_x)):
        dis = np.zeros(len(img2_x));

        x_shift = np.array((img1_x[i], img1_y[i]));
        for j in range(0,len(img2_x)):
            v2 = np.array((img2_x[j], img2_y[j]));
            dis[j] = np.linalg.norm(x_shift - v2);
        partner += [np.argmin(dis)];

    return img2_x[partner], img2_y[partner]


def Integrate(Xn,Yn ,Xpos, Ypos,NPartner=10):
    # Paramether
    # Xn : X-Position for integration point
    # Yn : Y-Position for integration point
    # Xpos: X-Position of Gradient data
    # Ypos: Y-Position of Gradient data
    
    P = NPartner
    import numpy as np

    # Index where Y-Component start in array "C"
    K = len(Xpos)
    
    C = np.zeros((2*K, len(Xn) ))
    
    #loop over all koordinates
    for i, (xk ,yk) in enumerate(zip(Xpos, Ypos)):
         
        # calculate distance to dots
        dX =  xk - Xn
        dY =  yk - Yn

        #get index of closest points
        distance = np.sqrt(np.square(dX) + np.square(dY))
        distance_idx = np.argsort(distance)

        # distance to closest points 
        dX = Xn[distance_idx[0:P]] - xk
        dY = Yn[distance_idx[0:P]] - yk

        # --- X Components ---
        # taylor series Matrix
        M = np.array([np.ones(len(dX)), 
                      dX,
                      dY,#])
                      (dX**2)/2, 
                      dX*dY,
                      (dX**2)/2])

        # "Zero" Array with the number of Taylor polinom
        V = np.zeros((np.size(M,0),1))
        V[1] = 1

        #M_pinv = np.linalg.pinv(M)

        #C[i,distance_idx[0:P]] = (M_pinv @ V).T
        C[i,distance_idx[0:P]] = np.linalg.lstsq(M,V,rcond=None)[0].T
        # "Zero" Array with the number of Taylor polinom
        V = np.zeros((np.size(M,0),1))
        V[2] = 1
        #C[i+K,distance_idx[0:P]] = (M_pinv @ V).T
        C[i+K,distance_idx[0:P]] = np.linalg.lstsq(M,V,rcond=None)[0].T

    empty_row = np.zeros((C.shape[0],1))
    empty_row[2] = 1
    
    D = np.append(C, empty_row, axis=1)
    
    return D



# calculate reference
img_lables = GetSeperationCV(img)
plt.imshow(img_lables)

x_ref,y_ref = getMomentumDIY(img_lables, img)
plt.plot(x_ref,y_ref,"ro")
plt.show()



# calculate one measurment
img2_lables = GetSeperation(img2)
x2, y2 = getMomentumDIY(img2_lables,img2)
ref = plt.plot(x_ref, y_ref, "ro", label='Referenze')
current = plt.plot(x2, y2, "bo", label='Change')

plt.legend()

plt.figure()
x2_sorted, y2_sorted = getPartner(x_ref,y_ref,x2,y2)

# gradient calculation
dx = x2_sorted - x_ref
dy = y2_sorted - y_ref


for idx,idy in zip(dx,dy):
    plt.plot([0,idx],[0,idy])
    
plt.figure()
plt.plot(dx)

plt.figure()
plt.plot(dy);



# calculate one measurment
# Genereate Mask
img2_lables = GetSeperation(img2)

# get centroid of dot section
x2, y2 = getMomentumDIY(img2_lables,img2)

# points may not be in the same order,
#match by minimal eucledian distance
%time x2_sorted, y2_sorted = getPartner(x_ref,y_ref,x2,y2)

# gradient calculation
dx = x2_sorted - x_ref
dy = y2_sorted - y_ref

# number of neighbouring points
P = 10

#1 construct vector Gvia Eq. (2)
G = np.concatenate([dx,dy])

#2 choose desired shape data locations [Xn,Yn]
n_res = 5
Future_samples_x = np.linspace(50,img.shape[0]-50,n_res)
Future_samples_y = np.linspace(50,img.shape[1]-50,n_res) 

Xn_2D, Yn_2D =  np.meshgrid(Future_samples_x, Future_samples_y)
Xn, Yn = Xn_2D.ravel(), Yn_2D.ravel()

D = Integrate(Xn, Yn, x2_sorted, y2_sorted,NPartner=10)

#Z = np.linalg.pinv(D) @ G
Z2 = np.linalg.lstsq(D,G,rcond=None)[0]

#plt.imshow(np.reshape(Z[0:-1],(n_res,n_res)))






%%timeit


plt.imshow(np.reshape(Z2[0:-1],(n_res,n_res)))

# # number of neighbouring points
# P = 10

# #1 construct vector Gvia Eq. (2)
# G = np.concatenate([dx,dy])

# #2 choose desired shape data locations [Xn,Yn]
# n_res = 6
# # Xn_array = np.linspace(50,img.shape[0]-50,n_res)
# # Yn_array = np.linspace(yk_array[0],yk_array[-1],n_res) 
# Future_samples_x = np.linspace(50,img.shape[0]-50,n_res)
# Future_samples_y = np.linspace(50,img.shape[1]-50,n_res) 


# Xn_2D, Yn_2D =  np.meshgrid(Future_samples_x, Future_samples_y)
# Xn, Yn = Xn_2D.ravel(), Yn_2D.ravel()

# # index where the Y array starts
# K = len(x2_sorted)


# C = np.zeros((2*K, len(Xn) ))
# #loop over all koordinates
# for i, (xk ,yk) in enumerate(zip(x2_sorted, y2_sorted)):
#      # calculate distance to dots
#     dX =  xk - Xn
#     dY =  yk - Yn
    
#     #get index of closest points
#     distance = np.sqrt(np.square(dX) + np.square(dY))
#     distance_idx = np.argsort(distance)
    
#     # distance to closest points 
#     dX = Xn[distance_idx[0:P]] - xk
#     dY = Yn[distance_idx[0:P]] - yk

#     # --- X Components ---
#     # taylor series Matrix
#     M = np.array([np.ones(len(dX)), 
#                   dX,
#                   dY,#])
#                   (dX**2)/2, 
#                   dX*dY,
#                   (dX**2)/2])

#     # "Zero" Array with the number of Taylor polinom
#     V = np.zeros((np.size(M,0),1))
#     V[1] = 1
    
#     M_pinv = np.linalg.pinv(M)
    
#     C[i,distance_idx[0:P]] = (M_pinv @ V).T
    
#     # "Zero" Array with the number of Taylor polinom
#     V = np.zeros((np.size(M,0),1))
#     V[2] = 1
#     C[i+K,distance_idx[0:P]] = (M_pinv @ V).T
     

# empty_row = np.zeros((C.shape[0],1))
# empty_row[2] = 1
# D = np.append(C, empty_row, axis=1)
# # D = C


D.shape

C.shape

np.append(C, empty_row, axis=1).shape

empty_row.shape

plt.imshow(D)



%time Z = np.linalg.pinv(D) @ G

np.reshape(Z,(n_res,n_res))

Z.shape

plt.imshow(np.reshape(Z[0:-1],(n_res,n_res)))

# Test - making momentcalculation by hand

k = img_lables.max()
k


lable_1d[i]

moment

%%timeit -n 1
k = img_lables.max() +1
moment = np.zeros(k)
img_1d = img.ravel()
lable_1d = img_lables.ravel()

for i in range(len(img_1d)):
    moment[lable_1d[i]] = img_1d[i] + moment[lable_1d[i]]

img_sub = img.copy().astype(np.float64)
mask = img_lables != 10
img_sub[mask] = np.NAN
plt.imshow(img_sub)


x_len = np.shape(img)[0]
y_len = np.shape(img)[1]

X = np.arange(x_len)[np.newaxis].T @ np.ones(y_len)[np.newaxis]
Y = (np.arange(y_len)[np.newaxis].T @ np.ones(x_len)[np.newaxis]).T
x0 = X**0 
y0 = Y**0

x1 = X**1 
y1 = Y**1


%time M00 = np.sum(img_sub)
print(M00)

%time M00 = np.sum(img.ravel() * x1.ravel())
print(M00)

%time M00 = np.sum(img_sub * y1)
print(M00)



```
